```python
import pandas as pd
import numpy as np
import seaborn as sns
import json
```

```python
#open saved files and preview
df_2000 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz", low_memory = False)
df_2000.head()
```

```python
df_2001 = pd.read_csv("Data/final_tmdb_data_2001.csv.gz", low_memory = False)
df_2001.head()
```

```python
#visualizations
ax = sns.barplot(data=df_2001, x='certification', y='revenue')
ax.set_title("2001 Movies Revenue Based on Rating");
```

```python
ax = sns.barplot(data=df_2000, x='certification', y='revenue')
ax.set_title("2000 Movies Revenue Based on Rating");
```

```python
#combine/concat dfs into one
#source: https://pandas.pydata.org/docs/user_guide/merging.html
frames = [df_2000, df_2001]

tmdb_results = pd.concat(frames)
```

```python
#review files merged
tmdb_results.head()
```

```python
tmdb_results.to_csv('Data/tmdb_results_combined.csv.gz', compression='gzip',index=False)
```

```python
#check df info
tmdb_results.info()
```

```python
#movies with values > 0 for budget OR revenue
#exclude any movies with 0's for budget AND revenue

filter_budget = tmdb_results['budget']>0
filter_revenue = tmdb_results['revenue']>0
tmdb_results.loc[filter_budget | filter_revenue]
```

```python
tmdb_results['certification'].value_counts()
```

```python
#movies in each of the certification categories (G/PG/PG-13/R)

filter_G = tmdb_results['certification']=='G'
filter_PG = tmdb_results['certification']=='PG'
filter_PG13 = tmdb_results['certification']=='PG-13'
filter_R = tmdb_results['certification']=='R'
```
